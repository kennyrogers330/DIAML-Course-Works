```python
# Question 1 Assignment_1_DIAML
"""
1. This Program aims to calculate the number of folds necessary for a 1mm thick paper to exceed the 
height 8848 m of the everest mountain
"""
def calc_folds(current_thickness, max_height):
    n = 0 # This counter will increase as the number of folds increases
    while current_thickness < max_height : # Each iteration will represent an additional fold of the paper
        current_thickness *= 2 # The current thickness shall be doubled on every fold
        n +=1 
    return n

# Initializing the thickness of the paper to 1 millimeter    
thickness = 1 #mm

# Initializing the height variable used as the height of Mountain Everest to 8848 meters
height = 8848 #m

#Converting the thickness of the paper to meter units 
thickness = 1 / 1000

# Calling the function calc_folds with arguments thickness and height 
folds = calc_folds(thickness, height)

#printing the final results 
print("\nA piece of paper, ", thickness, "mm thick, would exceed the height of Mount Everest in ", folds , " folds\n")
```

```python
# Question 2 Assignment_1_DIAML

#importing the math module 
import numpy as np

"""
This function aims to calculate the time required for the volume of water in a reservoir to decrease below its original value.
This time is refered to as "half-life" in exponential decay processes. 
"""
def calc_time_for_half_life(decay_rate):
    """ 
    > The half-life of an exponential decay equals to the negative of the natural logarithm of 
      2 divided by the decay rate. 
    
    > From the decay function v(t) = v(0)exp(-at) provided, the decay_rate is negative

    > Therefore this function will return the half life computed.
    """
    return -np.log(2) / (-decay_rate)
    
# Assigning the return value of the function to the variable 
half_life = calc_time_for_half_life(0.1)

"""
> Get the next time unit after the half life by adding the a single time unit to the half-life rounded up to two decimal points.

> This is done because we are asked the time required for the volume to fall below its original value yet the half life computes the time 
  required for a volume to fall to half its original value exactly.

> The single time units is one hundredth of a second because the half-life has two decimal points as well.
  The final result is rounded up to two decimal points as well. 
 
"""
time = round(round(half_life, 2) + 0.01, 2)

print("\nTime required for the volume to drop below half of the original value= ", time, "Time Units", end="\n")
```

```python
#Question 3 Assignment_1_DIAML 

#import the math module
import math

"""
 This function computes the total amount after compound interest on a deposit in a given time
"""
def calc_amount_after_compound_interest(p, r, frequency):
    
    # The rate value is divided by 100 
    rate = r / 100

    # This loop runs for 5 iterations to compute the compound interest for five years
    for i in range(1, 6):
        # Assign the amount for the year i to the variable amount
        amount = math.ceil(p * (1 + rate / frequency)**(frequency * i))
        # print the amount value for the year i 
        print(f"\nThe amount of money obtained after year {i} = $ {amount}")
    return
# Call the function to run at indentation 0
calc_amount_after_compound_interest(100, 5, 1)
```

```python
#Question 4 assignment_1_DIAML

# import the math module 
import math

#This function computes the monthly payment of loan with a given interest rate
def calc_monthly_payment_on_loan(ln, r):

    # The rate value is divided by 100 
    r = r / 100
    
    """
    This loop runs for three iterations to compute the monhtly payment required in case
    the loan is to be covered in one, two, or three years. 
    """
    for i in range(1, 4):
        """
        The variable n calculates the total number of payments that would be made depending
        on the number of years in which the loan will be paid
        """
        n = i*12

        """
        The montly payment is calculted and assigned in the variable monthly_payment
        in case the loan is to be paid in years equal to the current iteration
        """
        monthly_payment = math.ceil(ln*((r*(1+r)**n) / ((1+r)**n - 1)))
        
        # print the computed monthly payment 
        print(f"\nPayment necessary to pay back the loan in {i} year(s) = $ {monthly_payment} ")
    return
# Call the function to serve its purpose
calc_monthly_payment_on_loan(20000, 1)
```

```python
# Question 5 assignment_1_DIAML

# import the pyplot sub-module of matplotlib for plotting graphs  
import matplotlib.pyplot as plt

# import the numpy module  
import numpy as np

"""
This function plots each day elapsed against each cumulative frequency until break-even day
"""
def plot_results(x, y):

    """
    The two font variables font1, font2, below serve to style the font of the graph's title and labels respectively
    """
    titleFont = {'color':'#009933','size':18, 'family':'sans-serif'}
    labelsFont = {'color':'#990000','size':15, 'family':'sans-serif'}

    #Plot the list x against y using a pyplot line graph
    plt.plot(x, y)
    # Set the title of the graph
    plt.title("Profits made on an investment per day until the break even day\n", fontdict = titleFont)
    
    #set the label for the x-axis which plots the days
    plt.xlabel("time in days", fontdict = labelsFont)

    #set the label for the y-axis which plots the cumulative profit
    plt.ylabel("profit made in $", fontdict = labelsFont)

    #Determine the index of the highest value in the list containing cumulative profits 
    break_even_day = np.argmax(y)

    """ 
    plot a vertical line a that marks the 70th day on the graph, the color of the line will be 
    green, the line style is dashed, and the label is 'break even day'
    """
    plt.axvline(x=x[break_even_day] , color='green', linestyle = '--', label='break_even day')  
    
    """ 
    plot a horizontal line a that marks the Initial Investment the graph, the color of the line will be 
    blue, the line style is dashed, and the label is 'Initial Investment'
    """
    plt.axhline(100000, color='blue', linestyle = '--', label='Initial Investment')  

    #This line displays the labels on the graph
    plt.legend()

    """
    Annotate the break even point cordinate to to mark the break-even day of the
    investment. I used the annotate function of the pyplot sub-module
    """
    plt.annotate('Break-even day', xy=(x[break_even_day], y[break_even_day]), xytext=(x[break_even_day] + 1, 100000), fontsize=12) 
    #Show the graph
    plt.show
    
    """
    This function displays the number of days required for a business to attain the break even point
     - The first parameter is the principle business investment
     - The second parameter is the growth rate of customers 
     - The third parameter is the number of customers expected on day one
     - The fourth parameter is the Profit in $ made per customer
    """
def determine_break_even_day(pIV, rate, day1_customers, profit_per_customer):
    # Create a list called x that will hold each day elapsed until the break-even day
    x = []
    
    # Create a list called y that will hold each cummulative profit achieved until the break-even point
    y = []
    
    """
    Calculate the increment rate affecting both customers and the resultant profits made
    """
    increment_rate = (100 + rate) / 100
    
    """
    Action 1: 
    Calculate profits made on day one
    """
    profitN = day1_customers * profit_per_customer

    """
    Action 2: 
    assign the profit for day one to a variable that will record the cumulative profits in future
    """
    cumulativeP = profitN

    """
    Action 3: 
    Deduct the profit made from the initial investment
    """
    pIV -= profitN
    
    """
    Action 4: 
    Create a variable that will serve as a counter of days worked until the break-even day and assign 1 to it 
    """
    days = 1

    """
    Action 5: 
    append day elapsed to the days list (x)
    """
    x.append(days)

    """
    Action 6:
    append the cumulative profit to the cumulative profits list (y)
    """
    y.append(cumulativeP)

    """
    This loop repeats the actions from Action 1 to Action 6
    until the principal investment reaches or falls below zero
    """
    while pIV > 0:
        #Action 1
        profitN *= increment_rate
        #Action 2
        pIV -= profitN
        #Action 3
        days += 1
        #Action 4
        cumulativeP += profitN
        #Action 5
        x.append(days)
        #Action 6
        y.append(cumulativeP)
    
    """
    Call the function that plots the list of days elapsed against the list of cumulative profits made
    until the break-even day 
    """   
    plot_results(x, y)
    return days
    
days = determine_break_even_day(100000, 1, 100, 10)
print(f"\n\nIt will take {days} days to recover the investment made as shown in the graph below: ", end="\n\n")
```

```python
# Question 6 assignment_1_DIAML


import pandas as pd
from scipy.interpolate import interp1d
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# This function reads a dataset related to Ebola and makes estimation by linear interpolation
def estimate_by_interpolation(): 
    #Read the file of ebola deaths and cases
    dataset = pd.read_excel('ebola_download.xls')
    
    # Extract the dates column
    Z = dataset['Date']      
    """
    Create a list that will hold all datess in the range between the first date in the 
    data set and the last date in the dataset
    """    
    Alldates = []
    #Assign the first date in the dataset to a variable 
    date1 = Z[0]
    # Append the first date to the list of all dates
    Alldates.append(date1)
    """
    This loop will run until the last value in the dates list and shall leave the last date 
    in the dataset into this variable
    """
    for x in Z:
        dateL = x
    """
    This loop will run as long as date1 and dateL don't contain dates of equal value
    """
    while date1 != dateL:
        #Increment each date by exactly one day
        date1 += pd.DateOffset(days=1)
        #Append the current date to the list of dates
        Alldates.append(date1)
    """    
    Convert the list of all dates into a one dimensional array of dates with 
    indexes that can serve as a dataframe
    """
    Alldates = pd.Series(Alldates)
    
    """
    Merge the the Alldates dataframe with the provided dataset and make the resultant column contain all the values in 
    the Alldates dataframe
    """
    dataset = pd.merge(dataset, Alldates.rename('Date'), how='right', on='Date')
    #Interpolate the Noofdays
    dataset['Noofdays'] = dataset['Noofdays'].interpolate()
    #Interpolate the Diff
    dataset['Diff'] = dataset['Date'].diff()
    #Interpolate the Death
    dataset['Death'] = dataset['Death'].interpolate().round(1)
    #Interpolate the Cases
    dataset['Cases'] = dataset['Cases'].interpolate().round(1)
    
    # Assign new filename to a variable
    file = 'cleaned_dataset.csv'
    
    #This condition uses the os module to deletes the file if it exists and creates a new file or just creates the file if it didn't exist
    if os.path.exists(file):
        os.remove(file)
    #Write the cleaned dataframe to the new file
    dataset.to_csv('cleaned_dataset.csv', index=False)  
    
    #read the cleaned file into a dataframe
    dataset = pd.read_csv('cleaned_dataset.csv')
    
    #Convert the Date column from string data type to a time-stamp 
    dataset['Date'] = pd.to_datetime(dataset['Date'], format='%Y-%m-%d')
    #Extract the Cases column into List X
    X = dataset['Cases']
    #Extract the Death column into List Y
    Y = dataset['Death']
    #Extract the Date column into List Z
    Z = dataset['Date']

    # Set the desired width height of the graph before plotting 
    plt.figure(figsize=(12, 6))
    
    # Format the dates on the x-axis into YYYY/MM/DD
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))
    
    # Scale the dates to an interval of twenty days
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=20))
    
    # Graph the Cases aginst dates distribution with a dotted line style     
    plt.plot(Z, X, label='Cases distribution', linestyle='-')
    
    # Graph the Deaths aginst dates distribution with a dotted line style
    plt.plot(Z, Y, label='Deaths distribution', linestyle='-')
    
    # Create a List of thresholds that were exceeded during the Ebola Epidemic in 2014
    thresholds = [100, 500, 1000, 2000, 5000]

    """
    This nested loop runs until the last threshold is compared to all values in the Cases list and all values in the Deaths list.
    For each threshold in the list, a loop runs throughout the Cases list to determine when such a threshold was first exceeded and it breaks at that 
    point and a same loop runs throughout the Deaths to determine when such a threshold was exceeded in the deaths list
    """
    for threshold in thresholds:
        #Nested loop through the Cases list
        for case in X:
            #Check the current threshold was exceeded
            if case > threshold:
               #Record the index of the first value in Cases to exceed this threshold
               index = np.argmax(X == case)
               
               #Plot the point on the graph again as a way of marking the threshold exceeded. Use the index recorded to specify xy-coordinates
               plt.plot(Z[index], X[index], marker = 'o', mec = '#cc00cc', mfc = '#cc00cc')
               #Break out of the loop because we need one threshold exceeding instance per threshold
               break
        for death in Y:
            #Nested loop through the Cases list
            if death > threshold:
                #Record the index of the first value in Deaths to exceed this threshold
                index = np.argmax(Y == death)
                #Plot the point on the graph again as a way of marking the threshold exceeded. Use the index recorded to specify xy-coordinates
                plt.plot(Z[index], Y[index], marker = 'o', mec = '#339933', mfc = '#339933')
                #Break out of the loop because we need one threshold exceeding instance per threshold
                break
    
    titleFont = {'color':'#009933','size':18, 'family':'sans-serif'}
    labelsFont = {'color':'#990000','size':15, 'family':'sans-serif'}
    
    #Rotate and align the dates so that they fit on the x-axis 
    plt.gcf().autofmt_xdate()
    #Set the title of the graph
    plt.title('\nCases and Deaths count of Ebola in 2014', fontdict = titleFont)
    #Set the X-axis label
    plt.xlabel('Dates', fontdict = labelsFont)
    #Set the Y-axis label
    plt.ylabel('Count of people', fontdict = labelsFont)
    #Display the labels at the coordinates specified in the parameter
    plt.legend(bbox_to_anchor=(1.2, 1.15))
    #Display the graph
    plt.show()
#Call the function to the main execution    
estimate_by_interpolation()
```

```python
# Question 7 assignment_1_DIAML

import pandas as pd
import os
import numpy as np
# This function opens the file cleaned_dataset.csv
def open_file():
    #assign the file name to a variable 
    file = 'cleaned_dataset.csv'
    #use the exists() function of os to determine if the file exists before openning 
    if os.path.exists(file):
        #read the dataset using pandas
        dataset = pd.read_csv('cleaned_dataset.csv')
        #return the dataset
        return dataset
    else:
        #print the error message if the file doesn't exist
        print("File not found")
        return
        
# This function calculates the growthrate per day in the number of Ebola cases and deaths
def calc_growthRate():
    #call the open file function to read the dataset
    dataset = open_file()
    #extract the cases column into the List X
    X = dataset['Cases']
    #extract the Deaths column into the List Y
    Y = dataset['Death']
    #This list will be used as a collection of Cases' growthrate per day 
    cases_growth_rate = []
    #This list will be used as a collection of Deaths' growthrate per day
    death_count_growth_rate = []

    #This loop iterates for a number of times equal to the length of the list X
    for i in range(len(X) - 1):
        #The current value of the iteration is recorded as the today's cases 
        current_case = X[i]
        #The value at the next index is recorded as the next day's cases
        next_case = X[i+1]
        """
        A daily rate is computed by multiplying the difference between the cases for
        two consecutive days with one hundred and dividing by the present day's cases
        """
        daily_rate = ((next_case - current_case) * 100) / current_case
        #The daily rate is appended to the list of daily growth rates mentioned
        cases_growth_rate.append(daily_rate)
    
    #This loop iterates for a number of times equal to the length of the list Y
    for i in range(len(Y) - 1):
        #The current value of the iteration is recorded as the today's Deaths 
        current_death_count = Y[i]
        #The value at the next index is recorded as the next day's Deaths
        next_death_count = Y[i+1]
        """
        A daily rate is computed by multiplying the difference between the deaths for
        two consecutive days with one hundred and dividing by the present day's deaths
        """
        daily_death_count_rate = ((next_death_count - current_death_count) * 100) / current_death_count
        death_count_growth_rate.append(daily_death_count_rate)
    # print the growth rate of Ebola Cases which equals the sum of cases daily growth rates divied by the length of list holding cases growth rate   
    print(f"\n\nThe average daily growth-rate of Ebola cases in 2014: {sum(cases_growth_rate) / len(cases_growth_rate)} %")  
    # print the growth rate of Ebola Deaths which equals the sum of deaths daily growth rates divied by the length of list holding deaths growth rate 
    print (f"\n\nThe average daily growth-rate of deaths caused by Ebola in 2014: {sum(death_count_growth_rate) / len(death_count_growth_rate)} %")
#call the function to run in the main execution
calc_growthRate()    
```
