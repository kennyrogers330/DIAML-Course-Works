```python
# Question 1 DIAML_Assignment_2

"""The next two lines serve to import libraries used in this program, namely pandas and matplotlib"""
import pandas as pd
import matplotlib.pyplot as pt

"""This function plots a scatter plot of malnutrition prevelence against GDP per capita, having taken the two dataframes as paramaters"""
def plot_datasets(ds1, ds2):
    #This dictionary will hold all GDP per capita values for all years and all countries
    Gdp_ = {}

    #This dictionary will hold all malnutrition prevelance values for all years and all countries
    Prevelance_ = {}

    #This loop iterates through the datasets and assigns all GDP values and assigns them to the Gdp_ dictionary 
    for year in ds1.columns[4:]:
        #Assign Gdp value to dictionary with column name (year) as key and cell as value of the dictionary
        Gdp_[year]  = ds1[year]
    
    #This loop iterates through the datasets and assigns all Malnutrition prevelence values to the Prevelance_ dictionary 
    for year in ds2.columns[4:]:
        #Assign Prevelance value to dictionary with column name (year) as key and cell as value of the dictionary
        Prevelance_[year] = ds2[year]

    GDP_keys = Gdp_.keys()
    #Assign all the values of the Gdp_ dictionary 
    GDP_values = Gdp_.values()

    Prevelance_keys = Prevelance_.keys()
    #Assign all the values of the Prevelance_ dictionary 
    Prevelance_values = Prevelance_.values()

    #This dictionary holds the font styling attributes for the title of the scatter plot namely font color, size and family
    titleFont = {'color':'#009933','size':18, 'family':'sans-serif'}
    
    #This dictionary holds the font styling attributes for the labels of the scatter plot namely font color, size and family
    labelsFont = {'color':'#990000','size':15, 'family':'sans-serif'}  
    
    #This line sets up the width of the graph to 12 and the height to 6
    pt.figure(figsize=(12, 6))
    
    #This line sets the title of the graph 
    pt.title('\nScatter plot showing malnitrution prevelence against GDP per capita\n', fontdict = titleFont)
    
    #This line sets the label of the x-axis
    pt.xlabel('GDP per capita', fontdict = labelsFont)

    #This line set the label of the y-axis
    pt.ylabel('malnitrution prevelence', fontdict = labelsFont)

    #This line plots the scatter plot of GDP per capita against malnutrition prevelence 
    pt.scatter(GDP_values, Prevelance_values, c='blue')

    #This line set the legend of the graph
    pt.legend({'malnitrution prevelence against GDP per capita':'malnitrution prevelence against GDP per capita'})

    #This line plots the graph
    pt.show()

#This function expects a dataframe as parameter and extracts values in all cells for all years and all countries
def getValues_forAll_years(dataset):
    #This dictionary will hold all columns of the dataframe as keys and cells as values of the dictionary
    amount_ = {}
    #This loop extracts all values of the dataframe into the dictionary
    for year in dataset.columns[6:]:
        #Assign the values for a given year into the dictionary with column name as key
        amount_[year] = dataset[year]
    #Return the dictionary
    return amount_

"""This function plots a scatter plot of malnutrition prevelence against GDP per capita, having taken the two dataframes as paramaters
like the previous function does but this function takes a third parameter called required_regions which be used 
to distinguish the distribution into regions by use of a particular color marker for a particular region"""
def plot_datasets_by_region(GDP, Prevelance, required_regions):

    #This dictionary holds the font styling attributes for the title of the scatter plot namely font color, size and family
    titleFont = {'color':'#009933','size':18, 'family':'sans-serif'}

    #This dictionary holds the font styling attributes for the labels of the scatter plot namely font color, size and family
    labelsFont = {'color':'#990000','size':15, 'family':'sans-serif'} 
    
    #This line sets up the width of the graph to 12 and the height to 6
    pt.figure(figsize=(12, 6))

    #This line sets the title of the graph 
    pt.title('\nScatter plot showing regional malnitrution prevelence against GDP per capita\n', fontdict = titleFont)
    
    #This line sets the label of the x-axis
    pt.xlabel('GDP per capita', fontdict = labelsFont)

    #This line set the label of the y-axis
    pt.ylabel('malnitrution prevelence', fontdict = labelsFont)
    
    
    """
    This loop iterates through regions list and for each region name it extracts all GDP per capita and malnitrution prevelence values for all years but
    from a particular region reached on as the iteration goes on
    """
    for regn in required_regions:
        
        #This line set the condition of extraction of the GPD per capita values by region from the the GDP per capita dataset
        GDP_condition = GDP['Region'] == regn
        
        #This line set the condition of extraction of the malnitrution prevelance values by region from the the malnitrution prevelance dataset
        Prevelance_condition = Prevelance['Region'] == regn 
        
        #Store the values of GPD per capita in a particular region into a dataframe 
        regional_GDP = GDP[GDP_condition]

        #Store the values of malnitrution prevelance in a particular region into a dataframe 
        regional_prevelance = Prevelance[Prevelance_condition]

        #Call the function getValues_forAll_years mentioned on line 61 to extract GDP per capita for all years in the current region and store them in the dictionary 
        regional_GDP_dict_years = getValues_forAll_years(regional_GDP)

        #Call the function getValues_forAll_years mentioned on line 61 to extract malnitrution prevelence for all years in the current region and store them in the dictionary 
        regional_prevelance_dict_years = getValues_forAll_years(regional_prevelance)
        
        #Assign all values in regional_GDP_dict_years to a list to ease plotting
        X = regional_GDP_dict_years.values()

        #Assign all values in regional_prevelance_dict_years to a list to ease plotting
        Y = regional_prevelance_dict_years.values()

        #plot GDP Per capita for the current region against malnitrution prevelance 
        pt.scatter(X, Y, label = f'{regn}')
        #Set legend for each region to figure out in the key of the graph
        pt.legend(
            # {'Latin America & Caribbean': 'Latin America & Caribbean', 
            #        'South Asia':'South Asia', 'Sub-Saharan Africa':'Sub-Saharan Africa', 
            #        'Europe & Central Asia': 'Europe & Central Asia',
            #       'Middle East & North Africa': 'Middle East & North Africa',
            #       'East Asia & Pacific': 'East Asia & Pacific'}, 
                  bbox_to_anchor=(1, 1))
    #Display the scatterplot 
    pt.show()

"""This function plots a scatter plot of malnutrition prevelence against GDP per capita, having taken the two dataframes as paramaters
and a third parameter called income level which be used to distinguish the distribution into income levels by use of a particular color marker 
for a particular income level"""

def plot_datasets_by_income(GDP, Prevelance, required_income_levels):
    
    #This dictionary holds the font styling attributes for the title of the scatter plot namely font color, size and family
    titleFont = {'color':'#009933','size':18, 'family':'sans-serif'}

    #This dictionary holds the font styling attributes for the labels of the scatter plot namely font color, size and family
    labelsFont = {'color':'#990000','size':15, 'family':'sans-serif'}  

    #This line sets up the width of the graph to 12 and the height to 6
    pt.figure(figsize=(12, 6))

    #This line sets the title of the graph 
    pt.title('\nScatter plot showing malnitrution prevelence against GDP per capita basing on income levels\n', fontdict = titleFont)

    #This line sets the label of the x-axis
    pt.xlabel('GDP per capita', fontdict = labelsFont)

    #This line set the label of the y-axis
    pt.ylabel('malnitrution prevelence', fontdict = labelsFont)

    """
    This loop iterates through the income levels list passed as parameter and for each income level, it extracts all GDP per capita and malnitrution prevelence values for all years
    """    
    for income in required_income_levels:

        #This line set the condition of extraction of the GPD per capita values by income level from the the GDP per capita dataset passed as argument 
        GDP_condition = GDP['IncomeGroup'] == income

        #This line set the condition of extraction of the malnitrution prevelence values by income level from the the GDP per capita dataset passed as argument 
        Prevelance_condition = Prevelance['IncomeGroup'] == income
        
        #Store the values of GPD per capita by income levels into a dataframe 
        income_GDP = GDP[GDP_condition]

        #Store the values of malnitrution prevelence by income levels into a dataframe 
        income_prevelance = Prevelance[Prevelance_condition]
        
        #Call the function getValues_forAll_years mentioned on line 61 to extract GDP per capita for all years with the current income level and store them in the dictionary 
        income_GDP_across_years = getValues_forAll_years(income_GDP)

        #Call the function getValues_forAll_years mentioned on line 61 to extract malnitrution prevelence for all years with the current income level and store them in the dictionary 
        income_prevelance_across_years = getValues_forAll_years(income_prevelance)

        #Assign all values in income_GDP_across_years to a list to ease plotting
        X = income_GDP_across_years.values()

        #Assign all values in income_prevelance_across_years to a list to ease plotting
        Y = income_prevelance_across_years.values()

        #plot GDP Per capita for the current income level against malnitrution prevelance 
        pt.scatter(X, Y, label=f'{income}')
        #Set legend for each region to figure out in the key of the graph
        pt.legend(
            # {'High income': 'High income', 
            #        'Low income':'Low income', 
            #        'Lower middle income':'Lower middle income', 
            #        'Upper middle income': 'Upper middle income'}, 
                   
                   bbox_to_anchor=(1, 1))
    #Display the graph
    pt.show()

#Read the GDP Per capita of all countries across years in csv format using the read_csv function of pandas
GDP = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_5728786.csv", skiprows=4)

#Read the malnutrition prevelence of all countries across years in csv format using the read_csv function of pandas
Prevelance = pd.read_csv("API_SH.STA.MALN.ZS_DS2_en_csv_v2_5736088.csv", skiprows=4)

#drop the unnamed column on index 67 since it's useless and can potentially cause errors
GDP = GDP.drop(columns=['Unnamed: 67'])

#drop the unnamed column on index 67 since it's useless and can potentially cause errors
Prevelance = Prevelance.drop(columns=['Unnamed: 67'])

#Plot the first scatter plot of GDP per capita against malnitrution prevelence by calling the function on line 8
plot_datasets(GDP, Prevelance)

#Read the metadata file that contains the six regions and income levels that are of interest to the us for the problem we are solving
meta_data = pd.read_csv('Metadata_Country_API_NY.GDP.PCAP.CD_DS2_en_csv_v2_5728786.csv')

#Extract the regions column of the dataframe into a dataframe
regions = meta_data['Region']

#Insert this regions dataframe into the existing dataframe of GDP per Capita to map regions to the values of GDP per capita
GDP.insert(2, 'Region', regions)

#Insert this regions dataframe into the existing dataframe of malnitrution prevelence at the second index to map regions to the values of malnitrution prevelence 
Prevelance.insert(2, 'Region', regions)

#Extract the IncomeGroup column of the dataframe into a dataframe since this column will gives the income levels of different countries 
income_levels = meta_data['IncomeGroup']

#Insert this Income group dataframe into the existing dataframe of GDP per Capita to map income levels to the values of GDP per capita
GDP.insert(3, 'IncomeGroup', income_levels)

#Insert this Income group dataframe into the existing dataframe of malnitrution prevelence to map income levels to the values of malnitrution prevelence
Prevelance.insert(3, 'IncomeGroup', income_levels)

#Hard code the six required regions into a list
required_regions = ['Latin America & Caribbean', 'South Asia','Sub-Saharan Africa','Europe & Central Asia','Middle East & North Africa','East Asia & Pacific']

#Pass the list to the function as a third argument after the two datasets of GDP and malnitrution prevelence
plot_datasets_by_region(GDP, Prevelance, required_regions)

#Hard code the four income levels needed into a list
required_income_levels = ['High income', 'Low income', 'Lower middle income', 'Upper middle income']

#Pass the list of income levels to the function as a third argument after the two datasets of GDP and malnitrution prevelence
plot_datasets_by_income(GDP, Prevelance, required_income_levels)
```

```python
# Question 2 DIAML_Assignment_2

#import quandl as a library that provides the datasets we need in this case
import quandl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Read the API_Key file containing the my API key
quandl.read_key("./mkennyro_API_Key.txt")

#Read the wheat prices' dataset
Wheat_Prices = quandl.get('ODA/PWHEAMT_USD')

#Read the Crude oil prices' dataset
Crude_Oil_Prices = quandl.get('WGEC/WLD_CRUDE_WTI')

#Read the Gold prices' dataset
Gold_Prices = quandl.get('BUNDESBANK/BBK01_WT5511')

"""Since for each dataset the prices column is calles 'Value' I changed to make sure 
    on merging the three datasets I get three differently named columns.
"""
Wheat_Prices = Wheat_Prices.rename(columns={'Value' : 'Wheat'})
Crude_Oil_Prices = Crude_Oil_Prices.rename(columns={'Value' : 'Crude Oil'})
Gold_Prices = Gold_Prices.rename(columns={'Value' : 'Gold'})

"""I'm merging the three datasets so that only the dates that are existing in all three datasets remain in the merged dataset"""
dataset = pd.merge(Wheat_Prices, Crude_Oil_Prices, on = "Date", how = "inner").merge(Gold_Prices, on = 'Date', how = 'inner')

#Extract the wheat price values into a list
X = dataset['Wheat'].tolist()

#Extract the crude oil price values into a list
Y = dataset['Crude Oil'].tolist()

#Extract the Gold price values into a list
Z = dataset['Gold'].tolist()

#Extract the dates in the dataset to a list
D = dataset.index.tolist()

#This dictionary holds the font styling attributes for the title of the graph namely font color, size and family
titleFont = {'color':'#009933','size':18, 'family':'sans-serif'}

#This dictionary holds the font styling attributes for the labels of the graph namely font color, size and family
labelsFont = {'color':'#990000','size':15, 'family':'sans-serif'}

#Set the figure size
plt.figure(figsize=(12, 6))

plt.grid()

#This line sets the title of the graph 
plt.title('\nPrices of Wheat, Crude Oil, and Gold in $', fontdict = titleFont)

#This line sets the x-axis label of the graph 
plt.xlabel('Dates', fontdict = labelsFont)

#This line sets the y-axis label of the graph 
plt.ylabel('Prices', fontdict = labelsFont)

#Plot wheat prices against dates 
plt.plot(D, X, label='Wheat Prices')

#Plot Crude oil prices against dates 
plt.plot(D, Y, label='Crude Oil Prices')

#Plot Gold prices against dates 
plt.plot(D, Z, label='Gold Prices')

"""This loop iterates through the dataset columns namely Wheat, Crude Oil, Gold to determine the maximum and minimum values in each column and plots those two points"""
for val in dataset.columns:
    
    #Assgin the current column to a list
    list = dataset[val]

    #Assign the index of the maximum value in the list to a column
    index_max = np.argmax(dataset[val])

    #Assign the index of the minimum value in the list to a column 
    index_min = np.argmin(dataset[val])

    #Plot the minimum value in the current column and set a marker for the xy coordinates of the value
    plt.scatter(D[index_min], list.iloc[index_min], marker = 'o', label=f'minimum {val} price')

    #Plot the maximum value in the current column and set a marker for the xy coordinates of the value
    plt.scatter(D[index_max], list.iloc[index_max], marker = 'o', label=f'maximum {val} price')  

#Use the legend to set the position of the graph's key       
plt.legend(bbox_to_anchor=(1.5, 1))

#Display the graph
plt.show()
```

```python
# Question 4 DIAML_Assignment_2

#import the pandas library to read csv files 
import pandas as pd

#import the pyplot submodule of matplotlib to plot 2-D graphs
import matplotlib.pyplot as plt

#Import numpy used in the calculation of the CDF
import numpy as np

"""This function aims to plot the Cumulative distribution function, the mean, and median of a dataframe.
   It expects a series of the fertility distribution for a given year, the color to be assigned to the mean line graph and 
   a colour to be assigned to the median lin graph
"""
def getValues_forAll_years(year_series, mean_color, median_color):
    
    #Sort the data frame in ascending order
    x_year = np.sort(year_series)

    #Assign the probabilities of each value in the fertility dataframe to a list  
    y_year = np.arange(len(x_year)) / float(len(x_year))

    #Calcultate mean of the dataframe's distribution
    mean = year_series.mean()
   
    #Calculate the median of the dataframe's distribution 
    median = year_series.median()

    #Plot a vertical line to indicate the position of the mean of the distribution 
    plt.axvline(x=mean , linestyle = '--', color = f'{mean_color}',label=f'mean for the year {year_series.name}')  

    #Plot a vertical line to indicate the position of the median of the distribution 
    plt.axvline(x=median , linestyle = '--', color = f'{median_color}', label=f'median for the year {year_series.name}')  

    #Plot the list of probabilities of the data of distribution against the data of the distribution 
    plt.plot(x_year, y_year, label= f'CDF for the year {year_series.name}')     

#Read the CSV file of fertility rate downloaded from the world bank data
fertility = pd.read_csv('API_SP.DYN.TFRT.IN_DS2_EN_csv_v2_5729644.csv', skiprows=4)

#Read the CSV file of GDP Per capita downloaded from the world bank indicators
GDP = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_5728786.csv", skiprows=4)

#Extract only the distribution of GDP per capita in the year 2010 
GDP_2010 = GDP['2010']

#Extract only the distribution of fertility rate in the year 2010 
fertility_2010 = fertility['2010']

#This dictionary holds the font styling attributes for the title of the graph namely font color, size and family
titleFont = {'color':'#339933','size':18, 'family':'sans-serif'}

#This dictionary holds the font styling attributes for the labels of the graph namely font color, size and family
labelsFont = {'color':'#cc00cc','size':15, 'family':'sans-serif'}

#This dictionary holds the font styling attributes for the title of the graph namely font color, size and family
plt.figure(figsize=(15, 6))

#This dictionary holds the font styling attributes for the title of the graph namely font color, size and family
plt.title("\nFertility Rate against GDP per Capita for all countries in 2010\n", fontdict = titleFont)

#Set the label for the x-axis of the first graph
plt.xlabel("\nGDP per Capita", fontdict = labelsFont) 

#Set the label for the y-axis of the first graph
plt.ylabel("\nFertility Rate", fontdict = labelsFont)

#Plot the distribution of fertility rate against the GDP per capita in the year 2010
plt.scatter(GDP_2010, fertility_2010, c='green', label='Fertility rate vs GDP per capita')

#Add a legend to the scatter plot
plt.legend()

#Add a  grid to the scatter plot
plt.grid()

#Display the graph 
plt.show()

#Set the figure size of the second graph showing the Cumulative distribution function 
plt.figure(figsize=(15, 6))

#Set the title of the second graph showing the Cumulative distribution function 
plt.title("\nCumulative Distribution functions (CDFs) for the fertility rate variable using data from 1990 and 2010\n", fontdict = titleFont)
#Set the label for the x-axis of the second graph 
plt.xlabel("\nFertility Rate Variables", fontdict = labelsFont) 
#Set the label for the y-axis of the second graph 
plt.ylabel("\nProbability", fontdict = labelsFont)

#Add a  grid to the second graph
plt.grid()

#Call the function defined on line 16 to plot the CDF, mean and median of the distribution of fertility rate in the year 1990
getValues_forAll_years(fertility['1990'], "#8A2BE2", "#800000")
#Call the function defined on line 16 to plot the CDF, mean and median of the distribution of fertility rate in the year 2010
getValues_forAll_years(fertility['2010'], "#008080", "#FF0000")

#Set the legend for the second graph
plt.legend()

#Display the second graph
plt.show()
```
